```python
import json
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
#from nltk.corpus import stopwords
# from spacy.lang.en import English
# from spacy.lang.en.stop_words import STOP_WORDS
import re
from collections import Counter
```

```python
#获取预料
titles=[]
paragraphs=[]
qass=[]
contexts=[]
#解析SQuAD-2.0数据
with open("./dev-v2.0.json",'r',encoding='utf-8') as f:
    SQuAD=json.load(f)
    data=SQuAD["data"]
    print(len(data))
    for i in range(len(data)):
        try:
            titles.append(data[i]["title"])
            para=data[i]["paragraphs"]
            paragraphs.append(para)
            for j in range(len(para)):
                qas=para[j]["qas"]
                qass.append(qas)
                context=para[j]["context"]
                contexts.append(context)
        except AttributeError:
            print("没有字段")
```

```python
text=" ".join(contexts)
sentences = text.replace('\n'," ")
# sentences
#去除符号字符
r1 = u'[’!"#$%&\'()*+,-./:;<=>?@，。《》？“”‘’''！[\\]^_`{|}~]+'
context=re.sub(r1,'',sentences)
context
# # ses=ses.replace("  "," ")
```

```python
#词性还原
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
# lists=context.split(" ")
# new_lists=[]
# for word in lists:
#     new_lists.append(wnl.lemmatize(word))
# context=" ".join(new_lists)

# 分词
word_tokens=word_tokenize(context)
#获取停用词表
stop=open('3.0stopwords.txt','r+', encoding='utf-8')
filtered_sentence =[]
stopword = stop.read().split("\n")
for key in word_tokens:
#   小写 词形还原
    key=key.lower()
    key=wnl.lemmatize(key)
    if not(key.strip() in stopword) and (len(key.strip()) > 1):
        filtered_sentence.append(key)
sentences=" ".join(filtered_sentence)
#sentences
```

```python
#统计排序
word_dic=Counter(filtered_sentence)
```

```python
word_dic=dict(word_dic)
word_dic=sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
# word_dic.remove(word_dic[0])
# print((word_dic))
df=pd.DataFrame.from_dict(dict(word_dic),orient='index',columns=['totals'])
df=df.reset_index().rename(columns = {'index':'name'})
df.head(20)
```
