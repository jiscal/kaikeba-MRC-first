```python
import json
import pandas as pd
import numpy as np
import jieba
import re
from LAC import LAC
from collections import Counter
```

```python
#获取预料
titles=[]
paragraphs=[]
qass=[]
contexts=[]
#解析duread_robust数据
with open("./dev.json",'r',encoding='utf-8') as f:
    duread=json.load(f)
    data=duread["data"]
    print(len(data))
    for i in range(len(data)):
        try:
            titles.append(data[i]["title"])
            para=data[i]["paragraphs"]
            paragraphs.append(para)
            for j in range(len(para)):
                qas=para[j]["qas"]
                qass.append(qas)
                context=para[j]["context"]
                contexts.append(context)
        except AttributeError:
            print("没有字段")
```

```python
text=" ".join(contexts)
text
```

```python
#特殊字符
r1 = u'[’!"#$%&\'()〔〕*,:;、<>?@，。《》？“”‘’''！[\\]^_`{|}~【】くく★（）]+|[①②③⑤⑧⑨]'
texts=re.sub(r1,'',text)
#去除电话
t2=re.sub('\d*-\d*-\d*',"",texts)
#去除网址
t3=re.sub(r'[a-zA-Z]+//[^\s]*[.com|.cn]/[^\s]*/[^\s]*',"",t2)
#保留中文
t4=re.sub('[^\u4e00-\u9fa5]',"",t3)
t4
```

```python
#分词
lac = LAC(mode='seg')
word_tokens=lac.run(t4)
#去除停用词
stop=open('3.0stopwords.txt','r+', encoding='utf-8')
filtered_sentence =[]
stopword = stop.read().split("\n")
for key in word_tokens:
    if not(key.strip() in stopword) and (len(key.strip()) > 1):
        filtered_sentence.append(key)

word_dic=Counter(filtered_sentence)
word_dic=dict(word_dic)
word_dic=sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
# word_dic.remove(word_dic[0])
# print((word_dic))
df=pd.DataFrame.from_dict(dict(word_dic),orient='index',columns=['totals'])
df=df.reset_index().rename(columns = {'index':'name'})
df.head(20)
```
